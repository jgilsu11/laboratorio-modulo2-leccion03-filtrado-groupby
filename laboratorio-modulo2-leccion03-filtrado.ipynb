```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
# Carga el dataset de trabajo
netflix_concat= pd.read_csv("datos/netflix_concat.csv", index_col= 0)    #Contrl espacio para ver por donde voy
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
netflix_concat.head(2)
```

```python
filtro_lanzamiento= netflix_concat["type"] == "Movie"
filtro_release= netflix_concat["release_year"] == 2021
netflix_concat[filtro_lanzamiento & filtro_release].head(3)
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
filtro_duración= netflix_concat["duration"] == 90
netflix_concat[filtro_duración]["title"]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
filtro_año= netflix_concat["release_year"] > 2018
filtro_país= netflix_concat["country"] == "India"
netflix_concat[filtro_año & filtro_país]["title"]
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
filtro_clasifiacion= ["TV-14", "R"]
netflix_concat[netflix_concat["rating"].isin(filtro_clasifiacion)].filter(items= ["title","rating"])
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos ha dirigido cada uno de los directores?

```python
netflix_concat["director"].unique()
```

```python
netflix_concat["director"] == "Christopher Nolan"
```

```python
lista_director= ["Christopher Nolan", "Martin Scorsese"]
titulos_chris_martin= netflix_concat[netflix_concat["director"].isin(lista_director)]
titulos_chris_martin.value_counts().sum()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
netflix_concat["Language"].value_counts()
```

```python
lista_paises= ["United States", "United Kingdom", "Canada"]
netflix_US_UK_CA= netflix_concat[netflix_concat["country"].isin(lista_paises)]
netflix_US_UK_CA.value_counts() #Para ver que veamos que titulos son de esos paises


patron_regex= "/"                                             
más_language= netflix_US_UK_CA["Language"].str.contains(patron_regex, case= False, na= False)     #El case = False hace que se budque research con minuscula y con mayuscula si se pone case= True solo dará las minusculas
netflix_US_UK_CA[más_language].value_counts().sum()  #Para que veamos cuantos de esos títulos tienen más de un lenguaje
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
filtro_rango_año= netflix_concat["release_year"].between(2015, 2020, inclusive= "both")
netflix_concat[filtro_rango_año].filter(items= ["title","release_year"])
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
netflix_concat["IMDB Score"]
```

```python
filtro_IMDB= netflix_concat["IMDB Score"].between(6.5, 8, inclusive= "neither" )
filtro_duracion= netflix_concat["duration"].isna()
netflix_concat[filtro_IMDB & filtro_duracion].head(3)
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 o 150 minutos?

```python
patron_regex1= "^150" 
patron_regex2="^90"                                   
_noventa=netflix_concat["duration"].str.contains(patron_regex2, case= False, na= False, regex= True)
_cientocincuenta=netflix_concat["duration"].str.contains(patron_regex1, case= False, na= False, regex= True)
total= _cientocincuenta | _noventa
netflix_concat[total].head(3)
```

```python
_noventa.sum()
```
